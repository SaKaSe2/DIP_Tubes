# Minggu 7-8: Advanced Normalization (Slang to Formal)
Tahap ini merupakan kelanjutan dari preprocessing. Tujuannya adalah menormalisasi bahasa gaul (slang), singkatan, dan *typo* menjadi bahasa baku tanpa mengubah makna kalimat asli (Data Integrity).

```python
import pandas as pd
import re

# Memuat dataset hasil Basic Cleaning dari proses sebelumnya
df = pd.read_excel('dataset_cleaned.xlsx')
df.head(3)
```

### 1. Membangun Slang Dictionary
Kita definisikan kamus sederhana berisi pemetaan dari kata slang menjadi kata formal. Pada praktiknya kamus ini akan terus diperbarui.

```python
# Kamus translasi slang/bahasa gaul ke formal
slang_dict = {
    'aq': 'saya',
    'aku': 'saya',
    'gue': 'saya',
    'gw': 'saya',
    'bngung': 'bingung',
    'bgt': 'sekali',
    'banget': 'sekali',
    'mo': 'mau',
    'yg': 'yang',
    'dgn': 'dengan',
    'utk': 'untuk',
    'jd': 'jadi',
    'tp': 'tapi',
    'kalo': 'kalau',
    'kl': 'kalau',
    'klo': 'kalau',
    'aja': 'saja',
    'jgn': 'jangan',
    'udh': 'sudah',
    'udah': 'sudah',
    'gak': 'tidak',
    'ga': 'tidak',
    'g': 'tidak',
    'bisa': 'bisa',
    'laper': 'lapar',
    'healing': 'penyembuhan',
    'stress': 'stres',
    'bikin': 'membuat',
    'bener': 'benar',
    'tau': 'tahu',
    'taunya': 'tahunya',
    'kayak': 'seperti',
    'kyk': 'seperti',
    'emang': 'memang',
    'krn': 'karena',
    'karna': 'karena',
    'knetz': 'netizen korea',
    'indo': 'indonesia'
}

def normalize_slang(text, dictionary):
    if not isinstance(text, str):
        return text
    
    # Pisahkan kalimat berdasarkan spasi (tokenisasi sederhana)
    words = text.split()
    
    # Ganti kata jika ditemukan dalam kamus slang, jika tidak gunakan kata aslinya
    normalized_words = [dictionary.get(word, word) for word in words]
    
    return ' '.join(normalized_words)
```

### 2. Menguji Fungsi Normalisasi (Evaluasi Kasus Khusus)

```python
# Uji coba menggunakan contoh pada laporan
contoh_teks = 'aq bngung bgt mo nangis'
print('Data Mentah  :', contoh_teks)
print('Data Normal  :', normalize_slang(contoh_teks, slang_dict))
```

### 3. UTS (Evaluasi I): Demo Perbandingan Data Mentah vs Data Bersih
Menerapkan fungsi pada seluruh dataset untuk mengevaluasi Data Integrity (bahwa proses pembersihan tidak mengubah makna asli).

```python
# Terapkan normalisasi ke dataset
df['normalized_text'] = df['cleaned_text'].apply(lambda x: normalize_slang(x, slang_dict))

# Tampilkan sampel untuk pembuktian (Perbandingan Before - After)
pd.set_option('display.max_colwidth', None)
demo_df = df[['full_text', 'cleaned_text', 'normalized_text']].sample(10, random_state=42)

demo_df.rename(columns={
    'full_text': 'Teks Asli (Kotor)', 
    'cleaned_text': 'Basic Cleaned', 
    'normalized_text': 'Normalized (Formal)'
}, inplace=True)

demo_df
```

```python
# Simpan hasil akhir ke format Excel baru
df.to_excel('dataset_normalized.xlsx', index=False)
print('Dataset berhasil dinormalisasi dan disimpan ke dataset_normalized.xlsx!')
```
