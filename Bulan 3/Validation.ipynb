# Minggu 11-12: Validation & Documentation

Bagian ini buat ngambil sampel data 100 baris secara acak. Tujuannya biar kita bisa ngecek secara manual apakah hasil cleaning, normalisasi, dan labeling kita udah bener atau masih berantakan.

```python
import pandas as pd
import os

# Kita coba buka file hasil dari bulan 3 dulu, kalo ga ada pake file bulan 2
file_bulan_3 = 'dataset_labeled_bulan3.xlsx'
file_bulan_2 = '../Bulan 2/pipeline_akhir_bulan_2.xlsx'

if os.path.exists(file_bulan_3):
    df = pd.read_excel(file_bulan_3)
    print('Sip, pake data dari hasil labeling Bulan 3')
elif os.path.exists(file_bulan_2):
    df = pd.read_excel(file_bulan_2)
    print('Pake data pipeline Bulan 2 karena Bulan 3 belom dirun')
else:
    # Fallback ke basic cleaning kalo pipeline juga belom dirun
    df = pd.read_excel('../Bulan 2/dataset_cleaned.xlsx')
    print('Pake data basic cleaning')

print(f'Total data kita ada: {len(df)} baris')
```

### Ambil Sampel buat Validasi Manual
Kita ambil 100 data acak, terus tambahin kolom kosong biar nanti gampang ngecek dan ngisi bener/salahnya di Excel.

```python
# Ambil 100 baris acak (random_state biar hasil acaknya ga berubah-ubah terus)
sample_df = df.sample(n=min(100, len(df)), random_state=42).copy()

# Bikin kolom kosong buat tempat kita ngecek manual
sample_df['Pengecekan_Manual (Sesuai/Tidak)'] = ''
sample_df['Catatan_Revisi'] = ''

# Pilih kolom penting aja biar gampang diliat
kolom_pilihan = ['full_text']
if 'cleaned_text' in sample_df.columns: kolom_pilihan.append('cleaned_text')
if 'normalized_text' in sample_df.columns: kolom_pilihan.append('normalized_text')
if 'Kategori_Mental' in sample_df.columns: kolom_pilihan.append('Kategori_Mental')
kolom_pilihan.extend(['Pengecekan_Manual (Sesuai/Tidak)', 'Catatan_Revisi'])

sample_df = sample_df[kolom_pilihan]

display(sample_df.head())
```

```python
# Save ke excel biar gampang dibuka n divalidasi
output_name = 'sample_validasi_manual.xlsx'
sample_df.to_excel(output_name, index=False)
print(f'Done! File buat validasi udah disimpen namanya: {output_name}')
```
